```python
import numpy as np
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
#  cleaning the dataset ,converting
# totalcharges to numeric

df['totalcharges']=pd.to_numeric(df['totalcharges'],errors='coerce')
```

```python
#droping the missing rows

df = df.dropna()
```

```python
#encoding the traget variable

df['churn'] = df['churn'].map({'Yes':1,'No':0})
```

```python
# feature target saperation
X = df.drop(['customerid', 'churn'], axis=1)
y = df['churn']

print("Features (X) shape:", X.shape)
print("Target (y) shape:", y.shape)
```

```python
y.name,X.columns
```

```python
# feature and target separation

x = df.drop(['churn','customerid'],axis=1)
y = df['churn']

#here i removed the unnecessary column customer id too
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
cat_cols = x.select_dtypes(include='object').columns
num_cols = x.select_dtypes(exclude='object').columns
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train value counts:\n", y_train.value_counts(normalize=True))
print("y_test value counts:\n", y_test.value_counts(normalize=True))
```

```python
y.value_counts(normalize=True)
```

```python
y_train.value_counts(normalize = True)
```

```python
y_test.value_counts(normalize = True)
```

```python
# building sklearn encoder

preprocessor = ColumnTransformer(
    transformers=[
    ('cat',OneHotEncoder(drop='first',
    handle_unknown='ignore'),cat_cols),
    ('num','passthrough',num_cols) # Corrected typo here
    ]
)
```

```python
#now we will create a pipeline end to end for encoder one object , no leakage

from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
# pipeline = preprocessing + model together
# this avoid the means up in oredr

dt_pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier())
])
```

```python
# Train the decision tree
dt_pipeline.fit(x_train, y_train)
```

```python
# Evaluating the model

from sklearn.metrics import accuracy_score,classification_report
y_pred_dt = dt_pipeline.predict(x_test)

print(accuracy_score(y_test,y_pred_dt))
```

```python
print(classification_report(y_test,y_pred_dt))
```

```python
from sklearn.metrics import accuracy_score,classification_report
y_pred_dt = dt_pipeline.predict(x_test)
y_pred_train = dt_pipeline.predict(x_train)

print(accuracy_score(y_train,y_pred_train))
print(accuracy_score(y_test,y_pred_dt))


# if train acc is high , test acc is low - overfitting
# if train acc is low like(50),
# test acc is low - underfitting
# if train and test acc is good enough
# and same - properfitting
```

```python
# overfitting intuition
for depth in[2,4,6,8,10,None]:
  dt_pipeline=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',DecisionTreeClassifier(max_depth=depth,random_state=42))
  ])
  dt_pipeline.fit(x_train,y_train)
  train_acc=dt_pipeline.score(x_train,y_train)
  test_acc=dt_pipeline.score(x_test,y_test)
  print(f'depth:{depth},train={train_acc:.2f},Test={test_acc:.2f}')
```

```python
# feature importance

dt_model = dt_pipeline.named_steps['model']
encoder = dt_pipeline.named_steps['preprocessor']

features_names = encoder.get_feature_names_out()

importances = pd.Series(
    dt_model.feature_importances_, # Corrected to feature_importances_
    index = features_names
).sort_values(ascending=False)

importances.head(10)
# feature importance

```
